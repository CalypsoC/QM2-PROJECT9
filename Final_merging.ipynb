```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import requests
import io
```

```python
url1 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/Income.csv"
download = requests.get(url1).content
income = pd.read_csv(io.StringIO(download.decode('utf-8')))
income.head() 
```

```python
url2 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/edu.csv"
download = requests.get(url2).content
education = pd.read_csv(io.StringIO(download.decode('utf-8')))
education.head() 
```

```python
df = income.merge(education,on='STATE')
df.head()
```

```python
url3 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/Abortion strictness_raw.csv"
download = requests.get(url3).content
strictness = pd.read_csv(io.StringIO(download.decode('utf-8')))
strictness.head() 
```

```python
df =df.merge(strictness,on='STATE')
df.head()
```

```python
url4 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/political%20affilliation.csv"
download = requests.get(url4).content
politics = pd.read_csv(io.StringIO(download.decode('utf-8')))
politics.head() # we can see here we miss delete one column.
```

```python
politics=politics.drop(columns=['Unnamed: 0'])
politics.head()
```

```python
df =df.merge(politics,on='STATE')
df.head()
```

```python
url5 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/AGE_Raw.csv"
download = requests.get(url5).content
age = pd.read_csv(io.StringIO(download.decode('utf-8')))
age.head() 
```

```python
df =df.merge(age,on='STATE')
df.head()
```

```python
url6 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/Religiosity_Raw.csv"
download = requests.get(url6).content
religiosity = pd.read_csv(io.StringIO(download.decode('utf-8')))
religiosity.head() 
```

```python
df =df.merge(religiosity,on='STATE')
df.head()
```
