```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import requests
import io
```

```python
url1 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/rawdata/completingcollege_Raw.csv"
download = requests.get(url1).content
college = pd.read_csv(io.StringIO(download.decode('utf-8')))
college.head() 
```

cleaning for completing college 

```python
college = college.drop([0])
```

```python
college = college.replace('%', '', regex=True)
```

```python
college = college.rename(columns={'Name': 'STATE', '2016-2020': 'Completing college'})
```

```python
college.head()
```

```python
url2 = "https://raw.githubusercontent.com/CalypsoC/QM2-PROJECT9/main/rawdata/completing%20high%20school_raw.csv"
download = requests.get(url2).content
highschool = pd.read_csv(io.StringIO(download.decode('utf-8')))
highschool.head() 
```

```python
highschool = highschool.replace('%', '', regex=True)
```

```python
highschool = highschool.rename(columns={'Name': 'STATE', '2016-2020': 'Completing High school '})
```

```python
highschool.head()
```

merging these two dataframe to create eductaion.csv

```python
education_data = college.merge(highschool,on='STATE')
education_data
```

```python
from google.colab import files
education_data.to_csv('edu.csv', index=False)
```
